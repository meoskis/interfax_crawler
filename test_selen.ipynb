```python
def crawler(inn:int, delay=2):
    service = Service(executable_path=r"C:\Users\nkmeo\selenium\chromedriver-win64\chromedriver.exe")
    options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(service=service, options=options)
    
    driver.get(url=url)
    find_button = driver.find_element(By.XPATH, '//button[@class="btn btn-sm btn-toggler search-popup-link js-search-popup-link"]')
    find_button = find_button.click()
    time.sleep(delay)
    
    find_field = driver.find_element(By.XPATH, '//input[@class="search-form__input autocomplete-enabled js-search-input"]')
    find_field.send_keys(int(inn))
    time.sleep(delay)
    
    find_field = driver.find_element(By.XPATH, '//input[@class="search-form__input autocomplete-enabled js-search-input"]')
    find_field.send_keys(Keys.ENTER)
    time.sleep(delay)
    
    company_link = driver.find_element(By.XPATH, '/html/body/div[2]/div[2]/div/div/div/ul/li/div/div[2]/h3/a').get_attribute('href')
    driver.get(url=company_link)
    time.sleep(delay)
    
    interfax_name = driver.find_element(By.XPATH, '/html/body/div[3]/div[1]/div/div[1]/h1').text
    ogrn = driver.find_element(By.XPATH, '/html/body/div[3]/div[2]/div[2]/div[1]/div/div/div[1]/div/div[9]/div[2]').text
    kpp = driver.find_element(By.XPATH, '/html/body/div[3]/div[2]/div[2]/div[1]/div/div/div[1]/div/div[10]/div[2]').text
    
    driver.close()
    driver.quit()
    
    return(ogrn, kpp, interfax_name)
```

```python
data = pd.read_excel('Рейтинг крупнейших компаний России по объему реализации продукции RAEX-600.xlsx')
```

```python
clear_data = data[data['ИНН'].isna()]
clear_data['Название'] = clear_data['Название'].map(lambda x: x.replace('ё', 'е'))
clear_data['Название'].loc[346] = 'ООО Глобал Транс'
clear_data['Название'].iloc[304] = 'ООО Нордголд Менеджмент'
clear_data
```

```python
ogrn_filter = re.compile(r'ОГРН: [0-9]{13}')
inn_filter = re.compile(r'ИНН: [0-9]{10}')
kpp_filter = re.compile(r'КПП: [0-9]{9}')

def info_crawler(company_name, delay=2.5):
    service = Service(executable_path=r"C:\Users\nkmeo\selenium\chromedriver-win64\chromedriver.exe")
    options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(service=service, options=options)

    driver.get(url=url)
    find_field = driver.find_element(By.XPATH, '//*[@id="query"]')
    find_field.send_keys(company_name)
    time.sleep(delay)
    find_field.send_keys(Keys.ENTER)
    time.sleep(delay)

    i = 1
    while True:
        info = driver.find_element(By.XPATH, f'/html/body/div[1]/div[3]/div/div[1]/div[4]/div[{i}]/div[3]/div').text
        if ogrn_filter.findall(info) and inn_filter.findall(info) and kpp_filter.findall(info):
            ogrn = ogrn_filter.findall(info)[0][6:]
            inn = inn_filter.findall(info)[0][5:]
            kpp = kpp_filter.findall(info)[0][5:]
            break
        else:
            i+=1

    driver.close()
    driver.quit()
    
    return(ogrn, inn, kpp)
```

```python
ogrn_list = []
inn_list = []
kpp_list = []

for company in range(clear_data.shape[0]):
    ogrn, inn, kpp = info_crawler(clear_data.iloc[company]['Название'], delay=2)
    ogrn_list.append(ogrn)
    inn_list.append(inn)
    kpp_list.append(kpp)
```

```python
clear_data['ogrn'] = ogrn_list
clear_data['inn'] = inn_list
clear_data['kpp'] = kpp_list
```

```python
clear_data.drop(['ИНН'], axis=1, inplace=True)
```
